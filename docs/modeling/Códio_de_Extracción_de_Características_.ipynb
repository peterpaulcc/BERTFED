```python
import pandas as pd
from tsfresh import extract_features
from tsfresh.utilities.dataframe_functions import impute

# Carga tu archivo de datos
data = pd.read_excel('Final-Data.xlsx')


data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Inicializa un DataFrame para almacenar las características extraídas
extracted_features = pd.DataFrame()

# Itera sobre cada columna (serie de tiempo) y extrae características
for column in data.columns:
    # Prepara los datos en el formato requerido por tsfresh
    ts = data[column].reset_index()
    ts['id'] = column

    # Extrae características
    features = extract_features(ts, column_id='id', column_sort='Date')

    # Imputa para rellenar los posibles valores faltantes en las características extraídas
    imputed_features = impute(features)

    # Agrega las características extraídas al DataFrame principal
    extracted_features = pd.concat([extracted_features, imputed_features], axis=1)

# Muestra algunas de las características extraídas
print(extracted_features.head())
```

```python
# Guardar el DataFrame en un archivo CSV
extracted_features.to_csv('caracteristicas_extraidas.csv')
```
